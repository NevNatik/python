# Определение перспективного тарифа для телеком компании

## Описание проекта
Компания «Мегалайн» клиентам предлагают два тарифных плана: «Смарт» и «Ультра». 

Чтобы скорректировать рекламный бюджет, коммерческий департамент хочет понять, какой тариф приносит больше денег.
Необходимо сделать предварительный анализ тарифов на небольшой выборке клиентов. 

В нашем распоряжении данные 500 пользователей «Мегалайна»: кто они, откуда, каким тарифом пользуются, сколько звонков и сообщений каждый отправил за 2018 год. Нужно проанализировать поведение клиентов и сделать вывод — какой тариф лучше.

## Описание тарифов

1. Тариф «Смарт»
- Ежемесячная плата: 550 рублей
- Включено 500 минут разговора, 50 сообщений и 15 Гб интернет-трафика
- Стоимость услуг сверх тарифного пакета:
- минута разговора: 3 рубля
- сообщение: 3 рубля
- 1 Гб интернет-трафика: 200 рублей


2. Тариф «Ультра»

- Ежемесячная плата: 1950 рублей
- Включено 3000 минут разговора, 1000 сообщений и 30 Гб интернет-трафика
- Стоимость услуг сверх тарифного пакета:
- минута разговора: 1 рубль
- сообщение: 1 рубль
- 1 Гб интернет-трафика: 150 рублей

*Обратите внимание: «Мегалайн» всегда округляет вверх значения минут и мегабайтов. Если пользователь проговорил всего 1 секунду, в тарифе засчитывается целая минута.

## Описание данных

1. Таблица users (информация о пользователях):


- user_id — уникальный идентификатор пользователя
- first_name — имя пользователя
- last_name — фамилия пользователя
- age — возраст пользователя (годы)
- reg_date — дата подключения тарифа (день, месяц, год)
- churn_date — дата прекращения пользования тарифом (если значение пропущено, то тариф ещё действовал на момент выгрузки данных)
- city — город проживания пользователя
- tariff — название тарифного плана


2. Таблица calls (информация о звонках):


- id — уникальный номер звонка
- call_date — дата звонка
- duration — длительность звонка в минутах
- user_id — идентификатор пользователя, сделавшего звонок


3. Таблица messages (информация о сообщениях):


- id — уникальный номер сообщения
- message_date — дата сообщения
- user_id — идентификатор пользователя, отправившего сообщение


4. Таблица internet (информация об интернет-сессиях):


- id — уникальный номер сессии
- mb_used — объём потраченного за сессию интернет-трафика (в мегабайтах)
- session_date — дата интернет-сессии
- user_id — идентификатор пользователя


5. Таблица tariffs (информация о тарифах):


- tariff_name — название тарифа
- rub_monthly_fee — ежемесячная абонентская плата в рублях
- minutes_included — количество минут разговора в месяц, включённых в абонентскую плату
- messages_included — количество сообщений в месяц, включённых в абонентскую плату
- mb_per_month_included — объём интернет-трафика, включённого в абонентскую плату (в мегабайтах)
- rub_per_minute — стоимость минуты разговора сверх тарифного пакета (например, если в тарифе 100 минут разговора в месяц, то со 101 минуты будет взиматься плата)
- rub_per_message — стоимость отправки сообщения сверх тарифного пакета
- rub_per_gb — стоимость дополнительного гигабайта интернет-трафика сверх тарифного пакета (1 гигабайт = 1024 мегабайта)

## Шаг 2. Подготовьте данные. <a id='Подготовьте данные'> </a> 

````python
# удалить ошибочный столбец
internet_data = internet_data.drop(columns=['Unnamed: 0'], axis=1).reset_index()

# перевод из строкового типа в datatime
calls_data['call_date']= pd.to_datetime(calls_data['call_date'])
internet_data['session_date']= pd.to_datetime(internet_data['session_date'])
messages_data['message_date']= pd.to_datetime(messages_data['message_date'])

# создание столбца month для вывода значение месяца из даты
calls_data['month']= pd.DatetimeIndex(calls_data['call_date']).month
internet_data['month']= pd.DatetimeIndex(internet_data['session_date']).month
messages_data['month']= pd.DatetimeIndex(messages_data['message_date']).month

# Переименовать столбец tariff в таб.users_data. 
# ДЛя того чтобы можно обьединять с таб.tariffs_data, по одинаковому названию
users_data.rename(columns = {'tariff':'tariff_name'}, inplace = True)
````

````python
# Группировка  по столбцу user_id и month, и суммирование значений

calls_data= calls_data.pivot_table(index=['user_id','month'], values='duration', aggfunc = 'sum').reset_index()
for i, r in calls_data.iterrows():
    # минуты округляем в большую сторону
    calls_data.loc[i, 'duration'] = math.ceil(r['duration'])
display(calls_data.tail(5))
calls_data.info()

internet_data= internet_data.pivot_table(index=['user_id','month'], values='mb_used', aggfunc = 'sum').reset_index()
for i, r in internet_data.iterrows():
    # мегабайты округляем в большую сторону
    internet_data.loc[i, 'mb_used'] = math.ceil(r['mb_used'])
display(internet_data.head(5))
internet_data.info()

messages_data= messages_data.pivot_table(index=['user_id','month'], values='id', aggfunc = 'count').reset_index()
messages_data.rename(columns = {'id':'count_messages'}, inplace = True)
display(messages_data.head(5))
messages_data.info() 
````

````python
# Заполнением столбца tariff_name

# 1. Смотрим таблицу tariffs_data и его  типы данных
display(tariffs_data)
print(tariffs_data.dtypes)

# Переводим тип данных целое в вещественный тип
tariffs_data['minutes_included']= tariffs_data['minutes_included'].astype('float')
tariffs_data['messages_included']= tariffs_data['messages_included'].astype('float')
tariffs_data['mb_per_month_included']= tariffs_data['mb_per_month_included'].astype('float')

# 2. Функция для определения тарифа, 
# на подачу проверки подается 4 показателя из таблицы megaline_data
# название тарифного плана, длительность звонка в минутах, кол-во сообщений, объём потраченного за сессию интернет-трафика (в мегабайтах)
def tariff_null(tariff,duration_f,count_messages_f,mb_used_f):
    if pd.isnull(tariff):
        for i,r in tariffs_data.iterrows():
            if (duration_f <= r['minutes_included']):
                if (count_messages_f <= r['messages_included']):
                    if (mb_used_f <= r['mb_per_month_included']):
            # определяет какой тариф
                        return r['tariff_name']
                    else: # определяет какой тариф
                        return 'ultra'
    else:
        return tariff
               
# 3. Заполнением пустых значений с помощью вызова функции tariff_null
megaline_data['tariff_name']= megaline_data[['tariff_name' ,'duration', 'count_messages', 'mb_used']].apply(lambda x: tariff_null(x['tariff_name'],x['duration'],x['count_messages'],x['mb_used']),axis=1)
print('Кол-во пропущенных значений после обработки в tariff_name', megaline_data['tariff_name'].isnull().sum())
````

По всем важным полям в таблице провели обработку данных и изменны типы данных.

````python
# Посчитать помесячную выручку с каждого пользователя.
for ind, row in megaline_data.iterrows():
    # Проверка не было ли положенного по тарифу превышения кол-ва минут 
    if row['duration'] > row['minutes_included']:
        # в случае превышения разницу умножаем на стоимость минуты разговора сверх тарифного пакета
        a = (row['duration']-row['minutes_included'])*row['rub_per_minute']
    else:
        a = 0
    # Проверка не было ли положенного по тарифу превышения кол-ва сообщений 
    if row['count_messages'] > row['messages_included']:
        # в случае превышения разницу умножаем на стоимость отправки сообщения сверх тарифного пакета
        b= (row['count_messages']-row['messages_included'])*row['rub_per_message']
    else:
         b=0
    # Проверка не было ли положенного по тарифу превышения объёма интернет-трафика 
    if row['mb_used'] > row['mb_per_month_included']:
        # в случае превышения разницу умножаем на стоимость дополнительного гигабайта интернет-трафика сверх тарифного пакета
        c= ((row['mb_used']-row['mb_per_month_included'])/1024)*row['rub_per_gb']
    else:
        c=0
    # допольнительный столбец amount с расчетом помесячной выручки с каждого пользователя
    megaline_data.loc[ind, 'amount'] = a+b+c+row['rub_monthly_fee']
    
display(megaline_data.sample(10))
````

Компания «Мегалайн» клиентам предлагают два тарифных плана: «Смарт» и «Ультра».

Чтобы скорректировать рекламный бюджет, коммерческий департамент хочет понять, какой тариф приносит больше денег. Необходимо сделать предварительный анализ тарифов на небольшой выборке клиентов.

В нашем распоряжении данные 500 пользователей «Мегалайна»: кто они, откуда, каким тарифом пользуются, сколько звонков и сообщений каждый отправил за 2018 год. Нужно проанализировать поведение клиентов и сделать вывод — какой тариф лучше.

После обработки и подгтовки данных к анализу, получилась новая таблица с данными:

<h6>Описание таблицы:</h6>

- user_id -уникальный идентификатор пользователя
- month - месяц посчета услуг
- duration - длительность звонка в минутах
- age - возраст пользователя (годы)
- churn_date - дата прекращения пользования тарифом (если значение пропущено, то тариф ещё действовал на момент выгрузки данных)
- city — город проживания пользователя
- first_name — имя пользователя
- last_name — фамилия пользователя
- reg_date — дата подключения тарифа (день, месяц, год)
- tariff_name — название тарифного плана
- count_messages кол-во сообщений
- mb_used — объём потраченного за сессию интернет-трафика (в мегабайтах)
- rub_monthly_fee — ежемесячная абонентская плата в рублях
- minutes_included — количество минут разговора в месяц, включённых в абонентскую плату
- messages_included — количество сообщений в месяц, включённых в абонентскую плату
- mb_per_month_included — объём интернет-трафика, включённого в абонентскую плату (в мегабайтах)
- rub_per_minute — стоимость минуты разговора сверх тарифного пакета (например, если в тарифе 100 минут разговора в месяц, то со 101 минуты будет взиматься плата)
- rub_per_message — стоимость отправки сообщения сверх тарифного пакета
- rub_per_gb — стоимость дополнительного гигабайта интернет-трафика сверх тарифного пакета (1 гигабайт = 1024 мегабайта)

===============================================
Просчитала помесячную выручку с каждого пользователя.

Алгоритм вычисления выручки:

1. Проверка на превышение кол-во разговора в минутах в месяц и положенное кол-во минут по тарифному плану. В случае превышения, тогда мы прибавляем дополнительную стоимость минуты разговора сверх тарифного пакета. В случае не превышения, тогда прибыль от кол-ва звонка отсутствует;
2. Проверка на превышение кол-во сообщений в месяц и положенное кол-во сообщений по тарифному плану. В случае превышения, тогда мы прибавляем дополнительную стоимость отправки сообщения сверх тарифного пакета. В случае не превышения, тогда прибыль от кол-ва сообщений отсутствует;
2. Проверка на превышение потраченного за сессию интернет-трафика за месяц и положенного объёма интернет-трафика, включённого в абонентскую плату. В случае превышения, тогда мы прибавляем дополнительную стоимость дополнительного гигабайта интернет-трафика сверх тарифного пакета. Нужно учесть что дополнительный обьем считаеся в гигабайтах. В случае не превышения, тогда прибыль от объёма интернет-трафика отсутствует;
3. Создала дополнительный столбец amount, в котором делаем суммирование всех расчетов и плюс абонетская плата

Дисперсия и стандартное отклонение позволяют оценить степень колебания данных вокруг среднего значения.
