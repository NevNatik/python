### –í—ã–≤–æ–¥

–ü—Ä–æ–≤–µ–¥–µ–º –ø—Ä–µ–æ–±—Ä–∞–±–æ—Ç–∫—É —Å—Ç–æ–ª–±—Ü–æ–≤:

- ceiling_height

—Ç.–∫ –æ–Ω–∏ –Ω–∞–º –ø–æ–Ω–∞–¥–æ–±—è—Ç—Å—è –¥–ª—è –¥–∞–ª—å–Ω–µ–π—à–µ–≥–æ –∞–Ω–∞–ª–∏–∑–∞

–í—Å–µ –æ—Å—Ç–∞–ª—å–Ω—ã–µ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö –æ—Å—Ç–∞–≤–∏–º –±–µ–∑ –∏–∑–º–µ–Ω–µ–Ω–∏—è, —Ç.–∫. –æ–Ω–∏ –Ω–∏–∫–∞–∫–æ–π –≤ –∏—Å—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–∏ –æ –ø—Ä–æ–¥–∞–∂–µ –∫–≤–∞—Ä—Ç–∏—Ä –Ω–µ –∏–≥—Ä–∞—é—Ç –≤–∞–∂–Ω–æ —Ä–æ–ª–∏.

```python
# –†–∞–±–æ—Ç–∞ —Å –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –≤ —Å—Ç–æ–ª–±—Ü–µ ceiling_height (–≤—ã—Å–æ—Ç–∞ –ø–æ—Ç–æ–ª–∫–æ–≤ (–º))

#–ø–µ—Ä–µ–≤–µ–¥–µ–º –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–∞ –≤ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–π —Ç–∏–ø
real_estate_data['ceiling_height']= pd.to_numeric(real_estate_data['ceiling_height'])

print('–ö–æ–ª-–≤–æ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ ceiling_height', real_estate_data['ceiling_height'].isnull().sum())

#–ø—Ä–æ–≤–µ–ª–∞ –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫—É –ø–æ –Ω–∞—Å–µ–ª–µ–Ω–Ω–æ–º—É –ø—É–Ω–∫—Ç—É, –≤—Å–µ–≥–æ —ç—Ç–∞–∂–µ –≤ –¥–æ–º–µ –∏ –≤—ã—Å–æ—Ç–∞ –ø–æ—Ç–æ–ª–∫–æ–≤
for i, r in real_estate_data.groupby(['locality_name','floors_total','ceiling_height']):
    for row_index, row in r.iterrows():
        ln = row['locality_name']
        ft = row['floors_total']
        # –ò—â—É –ø—É—Å—Ç—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Å—Ç–æ–ª–±—Ü–µ ceiling_height, –ø–æ –Ω–∞—Å–µ–ª–µ–Ω–Ω–æ–º—É –ø—É–Ω–∫—Ç—É –∏ –∫–æ–ª-–≤–æ —ç—Ç–∞–∂–µ–π
        for i2,r2 in real_estate_data.query('ceiling_height.isnull() and locality_name == @ln and floors_total == @ft').iterrows():
            #–∑–∞–ø–∏—Å—ã–≤–∞—é –≤—ã—Å–æ—Ç—É —Å –æ–¥–∏–Ω–∞–∫–æ–≤—ã–º –Ω–∞—Å–µ–ª–µ–Ω–Ω—ã–º –ø—É–Ω–∫—Ç–æ–º –∏ –∫–æ–ª-–≤–æ —ç—Ç–∞–∂–µ–π
            real_estate_data.loc[i2, 'ceiling_height'] = row['ceiling_height']
            
#–û—Å—Ç–∞–ª—å–Ω—É—é —á–∞—Å—Ç—å –ø—Ä–æ–ø—É—Å–∫–æ–≤ –∑–∞–ø–æ–ª–Ω—è—é —Å—Ä–µ–¥–Ω–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ–º –ø–æ –≤—ã—Å–æ—Ç–µ –ø–æ—Ç–æ–ª–∫–æ–≤
med_ceiling_height = real_estate_data['ceiling_height'].median()
print('–°—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ—Ç–æ–ª–∫–æ–≤:', med_ceiling_height)

#–ü–æ–ª—É—á–µ–Ω–Ω—É—é –∑–Ω–∞—á–µ–Ω–∏–µ –ø—Ä–æ–ø–∏—à–µ–º –≤ –ø—Ä–æ–ø—É—Å–∫–∞—Ö –≤ —Å—Ç–æ–ª–±—Ü–µ 'ceiling_height'
real_estate_data['ceiling_height']= real_estate_data['ceiling_height'].fillna(med_ceiling_height)

print('–ö–æ–ª-–≤–æ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ—Å–ª–µ –æ–±—Ä–∞–±–æ—Ç–∫–∏ –≤ ceiling_height', real_estate_data['ceiling_height'].isnull().sum())
```

1. –ì—Ä—É–ø–ø–∏—Ä—É—é –¥–∞–Ω–Ω—ã–µ –Ω–∞—Å–µ–ª–µ–Ω–Ω—ã–µ –ø—É–Ω–∫—Ç—ã —Å –∫–æ–ª-–≤–∞–º–∏ —Ç–∞–∂–µ–π –∏ –≤—ã—Å–æ—Ç–∞ –ø–æ—Ç–æ–ª–∫–æ–≤. –≠—Ç–æ –Ω–∞–º –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –¥–ª—è —Ç–æ–≥–æ —á—Ç–æ–±—ã –Ω–∞–π—Ç–∏ –ø–æ—Ö–æ–∂–∏–µ–º –¥–æ–º–∞ —Å –æ–¥–∏–Ω–∞–∫–æ–≤—ã–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏. 
2. –°–æ–∑–¥–∞–µ–º —Å—Ä–µ–∑ –≥–¥–µ –æ—Ç—Å—É—Ç—Å–≤—É–µ—Ç –≤—ã—Å–æ—Ç–∞ –ø–æ—Ç–æ–ª–∫–æ–≤, –Ω–æ –Ω–∞—Å–µ–ª–µ–Ω–Ω—ã–π –ø—É–Ω–∫—Ç –∏ —ç—Ç–∞–∂–Ω–æ—Å—Ç—å —Å–æ–≤–ø–∞–¥–∞–µ—Ç.
3. –ü—Ä–æ—Å—Ç–∞–≤–ª—è—é –≤ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–∞–ø–∏—Å—è—Ö –∑–Ω–∞—á–µ–Ω–∏—è –∏–∑ –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∏
4. –û—Å—Ç–∞–≤—à–∏–µ—Å—è –∑–∞–ø–∏—Å–∏ –∫–æ—Ç–æ—Ä—ã–µ –Ω–µ –Ω–∞—à–ª–∏—Å—å –ø–æ –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–µ —è –∑–∞–ø–æ–ª–Ω–∏–ª–∞ –∑–Ω–∞—á–µ–Ω–∏–µ–º –º–µ–¥–∏–∞–Ω—ã.

–°–ª–µ–¥—É—é—â–∏–π —ç—Ç–∞–ø –æ–±—Ä–∞–±–æ—Ç–∫–∏ —ç—Ç–æ –≤—ã–≤–æ–¥ –≥—Ä–∞—Ñ–∏–∫–∞ —Ä–∞–∑–º–∞—Ö–∞ –∏ –ø–æ–ª—É—á–µ–Ω–∏–µ –Ω–æ—Ä–º–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π.

```python
#–°—Ç—Ä–æ—é –≥—Ä–∞—Ñ–∏–∫ —Ä–∞–∑–º–∞—Ö–∞  –¥–ª—è —Å—Ç–æ–ª–±—Ü–∞ ceiling_height
plt.ylim(real_estate_data['ceiling_height'].min(), real_estate_data['ceiling_height'].max())
bp = plt.boxplot(real_estate_data['ceiling_height'])
plt.show()
# –ü–æ–ª—É—á–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –∫–≤–∞–Ω—Ç–∏–ª–∏–π Q1 –∏ Q3
Q1 = bp['whiskers'][0].get_ydata()[1]
Q3 = bp['whiskers'][1].get_ydata()[1]
print('Q1-', Q1)
print('Q3-', Q3)

# –ü–µ—Ä–µ—Å–æ–∑–¥–∞—é —Ç–∞–±–ª–∏—Ü—É —Å –Ω–æ—Ä–º–∞–ª—å–Ω—ã–º–∏ –∑–Ω–∞—á–Ω–µ–∏—è–º–∏ –≤ —Å—Ç–æ–ª–±—Ü–µ  ceiling_height –≤—Ö–æ–¥—è—â–∏–µ –≤ –æ—Ç—Ä–µ–∑–æ–∫ Q1 –∏ Q3
real_estate_data= real_estate_data.query('(ceiling_height>=@Q1)&(ceiling_height<=@Q3)')

real_estate_data.info()
```

2.  –ò–∑–º–µ–Ω–∏—Ç—å —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö –≤ int. –°–Ω–∞—á–∞–ª–æ —Å—Ç—Ä–æ–∫–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–µ—Ä–µ–≤–æ–¥–∏–º –≤–æ float, –ø–æ—Ç–æ–º –ø–µ—Ä–µ–≤–æ–¥–∏–º –≤ —Ç–∏–ø int.

```python
# –†–∞–±–æ—Ç–∞ —Å –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –≤ —Å—Ç–æ–ª–±—Ü–µ floors_total (–≤—Å–µ–≥–æ —ç—Ç–∞–∂–µ–π –≤ –¥–æ–º–µ)
print('–ö–æ–ª-–≤–æ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ floors_total', real_estate_data['floors_total'].isnull().sum())
# –ø—Ä–æ—Å—Ç–∞–≤–∏–º 0 —Ç–∞–º –≥–¥–µ –Ω–µ —É–∫–∞–∑–∞–Ω–æ –∫–æ–ª-–≤–æ —ç—Ç–∞–∂–µ–π –≤ –¥–æ–º–µ
real_estate_data['floors_total'] = real_estate_data['floors_total'].fillna('1.0')
print('–ö–æ–ª-–≤–æ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ—Å–ª–µ –æ–±—Ä–∞–±–æ—Ç–∫–∏ –≤ floors_total', real_estate_data['floors_total'].isnull().sum())

#–ø–µ—Ä–µ–≤–µ–¥–µ–º –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–∞ –≤ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–π —Ç–∏–ø
real_estate_data['floors_total']= pd.to_numeric(real_estate_data['floors_total'])
#–ø–µ—Ä–µ–≤–µ–¥–µ–º –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–∞ –≤ —Ü–µ–ª–æ—á–∏—Å–ª–µ–Ω–Ω—ã–π —Ç–∏–ø
real_estate_data['floors_total'] = real_estate_data['floors_total'].astype('int')
```

### –í—ã–≤–æ–¥

–û–±—Ä–∞–±–æ—Ç–∫–∞ —Å—Ç–æ–ª–±—Ü–∞ floors_total:

1.  –ü—Ä–æ—Å—Ç–∞–≤–∏—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ 1 —Ç–∞–º, –≥–¥–µ –ø—Ä–æ–ø—É—â–µ–Ω–Ω–æ. –°–∫–æ—Ä–µ–µ –≤—Å–µ–≥–æ –æ–ø–µ—Ä–∞—Ç–æ—Ä –Ω–µ —É–∫–∞–∑–∞–ª –∫–æ–ª-–≤–æ —ç—Ç–∞–∂–µ–π, —Ç.–∫ –∫–≤–∞—Ä—Ç–∏—Ä–Ω—ã–π –¥–æ–º —Å –æ–¥–Ω–∏–º —ç—Ç–∞–∂–æ–º. 
2.  –ò–∑–º–µ–Ω–∏—Ç—å —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö –≤ int. –°–Ω–∞—á–∞–ª–æ —Å—Ç—Ä–æ–∫–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–µ—Ä–µ–≤–æ–¥–∏–º –≤–æ float, –ø–æ—Ç–æ–º –ø–µ—Ä–µ–≤–æ–¥–∏–º –≤ —Ç–∏–ø int.


```python
#–ø–µ—Ä–µ–≤–µ–¥–µ–º –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤ total_area, living_area, kitchen_area, cityCenters_nearest –≤ –≤–µ—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–π —Ç–∏–ø
real_estate_data['total_area']= pd.to_numeric(real_estate_data['total_area'])
real_estate_data['living_area']= pd.to_numeric(real_estate_data['living_area'])
real_estate_data['kitchen_area']= pd.to_numeric(real_estate_data['kitchen_area'])
real_estate_data['cityCenters_nearest']= pd.to_numeric(real_estate_data['cityCenters_nearest'])

# –†–∞–±–æ—Ç–∞ —Å –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –≤ —Å—Ç–æ–ª–±—Ü–µ living_area (–∂–∏–ª–∞—è –ø–ª–æ—â–∞–¥—å –≤ –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã—Ö –º–µ—Ç—Ä–∞—Ö(–º¬≤))
# –ù–∞—Ö–æ–¥–∏–º –∫–æ–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç —Å—Ä–µ–¥–Ω–µ–µ –æ—Ç–Ω–æ—à–µ–Ω–∏–µ –∂–∏–ª–æ–π –ø–ª–æ—â–∞–¥–∏ –∫ –æ–±—â–µ–π –ø–æ –≤—Å–µ–π —Ç–∞–±–ª–∏—Ü–µ
k = real_estate_data['living_area'].mean()/real_estate_data['total_area'].mean()
print('–ö–æ–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∂–∏–ª–æ–π –ø–ª–æ—â–∞–¥–∏ ',k)
print('–ö–æ–ª-–≤–æ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ living_area', real_estate_data['living_area'].isnull().sum())
# –¢–∞–º –≥–¥–µ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –¥–µ–ª–∞–µ–º —Ä–∞—Å—á–µ—Ç –∂–∏–ª–æ–π –ø–ª–æ—â–∞–¥–∏ = –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç * –Ω–∞ –æ–±—â—É—é –ø–ª–æ—â–∞–¥—å –∫–≤–∞—Ä—Ç–∏—Ä—ã
real_estate_data['living_area'] = real_estate_data['living_area'].fillna(k*real_estate_data['total_area'])
print('')

# –†–∞–±–æ—Ç–∞ —Å –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –≤ —Å—Ç–æ–ª–±—Ü–µ kitchen_area (–ø–ª–æ—â–∞–¥—å –∫—É—Ö–Ω–∏ –≤ –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã—Ö –º–µ—Ç—Ä–∞—Ö (–º¬≤))
# –ù–∞—Ö–æ–¥–∏–º —Å—Ä–µ–¥–Ω–µ–µ –æ—Ç–Ω–æ—à–µ–Ω–∏–µ –ø–ª–æ—â–∞–¥–∏ –∫—É—Ö–Ω–∏ –∫ –æ–±—â–µ–π –ø–æ –≤—Å–µ–π —Ç–∞–±–ª–∏—Ü–µ
k_kitchen = real_estate_data['kitchen_area'].mean()/real_estate_data['total_area'].mean()
print('–ö–æ–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –ø–ª–æ—â–∞–¥–∏ –∫—É—Ö–Ω–∏ ',k_kitchen)
print('–ö–æ–ª-–≤–æ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ kitchen_area', real_estate_data['kitchen_area'].isnull().sum())
# –¢–∞–º –≥–¥–µ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –¥–µ–ª–∞–µ–º —Ä–∞—Å—á–µ—Ç –ø–ª–æ—â–∞–¥–∏ –∫—É—Ö–Ω–∏ = –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç * –Ω–∞ –æ–±—â—é—é –ø–ª–æ—â–∞–¥—å –∫–≤–∞—Ä—Ç–∏—Ä—ã
real_estate_data['kitchen_area'] = real_estate_data['kitchen_area'].fillna(k_kitchen*real_estate_data['total_area'])
print('')

# –†–∞–±–æ—Ç–∞ —Å –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –≤ —Å—Ç–æ–ª–±—Ü–µ lcityCenters_nearest (—Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –¥–æ —Ü–µ–Ω—Ç—Ä–∞ –≥–æ—Ä–æ–¥–∞ (–º))
print('–ö–æ–ª-–≤–æ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ cityCenters_nearest', real_estate_data['cityCenters_nearest'].isnull().sum())
# –¢–∞–º –≥–¥–µ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –¥–µ–ª–∞–µ–º –∑–∞–º–µ–Ω—É –Ω–∞ —Å—Ä–µ–¥–Ω–µ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –¥–æ —Ü–µ–Ω—Ç—Ä–∞ –≥–æ—Ä–æ–¥–∞
real_estate_data['cityCenters_nearest'] = real_estate_data['cityCenters_nearest'].fillna(real_estate_data['cityCenters_nearest'].mean())
print('')

print('–ö–æ–ª-–≤–æ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ—Å–ª–µ –æ–±—Ä–∞–±–æ—Ç–∫–∏ –≤ living_area', real_estate_data['living_area'].isnull().sum())
print('–ö–æ–ª-–≤–æ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ—Å–ª–µ –æ–±—Ä–∞–±–æ—Ç–∫–∏ –≤ kitchen_area', real_estate_data['kitchen_area'].isnull().sum())
print('–ö–æ–ª-–≤–æ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ—Å–ª–µ –æ–±—Ä–∞–±–æ—Ç–∫–∏ –≤ cityCenters_nearest', real_estate_data['cityCenters_nearest'].isnull().sum())
```

### –®–∞–≥ 3. –ü–æ—Å—á–∏—Ç–∞–π—Ç–µ –∏ –¥–æ–±–∞–≤—å—Ç–µ –≤ —Ç–∞–±–ª–∏—Ü—É

```python
#–¶–µ–Ω–∞ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞ —Å –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ–º
real_estate_data['price_kv'] = round(real_estate_data['last_price']/real_estate_data['total_area'])

display(real_estate_data.head(5))
```

### –í—ã–≤–æ–¥

–ü—Ä–æ–≤–µ–ª–∞ —Ä–∞—Å—á–µ—Ç —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞ –∏ –¥–æ–±–∞–≤–∏–ª–∞ —Ä–∞—Å—á–µ—Ç –≤ —Å—Ç–æ–ª–±–µ—Ü price_kv

–°—Ç–æ–∏–º–æ—Å—Ç–∏ 1–∫–≤.–º = –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ(–¶–µ–Ω–∞/–ü–ª–æ—â–∞–¥—å –∫–≤–∞—Ä—Ç–∏—Ä—ã (–∫–≤.–º))






```python
#–î–æ–±–∞–≤–∏–º —Å—Ç–æ–ª–±–µ—Ü "–î–µ–Ω—å –Ω–µ–¥–µ–ª–∏"

#–§—É–Ω–∫—Ü–∏—è –ø–µ—Ä–µ–≤–æ–¥–∞ —Ü–∏—Ñ—Ä—ã –¥–Ω—è –Ω–µ–¥–µ–ª–∏ –≤ –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ
def weeekday(day):
    d = {0: '–ü–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫', 1:'–í—Ç–æ—Ä–Ω–∏–∫', 2:'–°—Ä–µ–¥–∞',3:'–ß–µ—Ç–µ—Ä–≥',4:'–ü—è—Ç–Ω–∏—Ü–∞',5:'–°—É–±–±–æ—Ç–∞', 6:'–í–æ—Å–∫—Ä–µ—Å–µ–Ω—å–µ'}
    for key in d:
        if day == key:
             return d[key]
#–ø–µ—Ä–µ—Ö–æ–¥ –ø–æ —Å—Ç—Ä–æ–∫–∞–º –∏ –∑–∞–ø–∏—Å—å –≤ —Å—Ç–æ–ª–±–µ—Ü            
for index,row in real_estate_data.iterrows():
    real_estate_data.loc[index, 'day_week'] = weeekday(row['first_day_exposition'].weekday()) 
display(real_estate_data.head(5))
```

```python
#–î–æ–±–∞–≤–∏–º —Å—Ç–æ–ª–±—Ü—ã "–º–µ—Å—è—Ü", "–≥–æ–¥"

#–§—É–Ω–∫—Ü–∏—è –ø–µ—Ä–µ–≤–æ–¥–∞ —Ü–∏—Ñ—Ä—ã –º–µ—Å—è—Ü–∞ –≤ –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ
def month_naim(mth):
    d = {1: '–Ø–Ω–≤–∞—Ä—å', 2:'–§–µ–≤—Ä–∞–ª—å', 3:'–ú–∞—Ä—Ç', 4:'–ê–ø—Ä–µ–ª—å', 5:'–ú–∞–π', 6:'–ò—é–Ω—å', 7:'–ò—é–ª—å', 8:'–ê–≤–≥—É—Å—Ç', 9:'–°–µ–Ω—Ç—è–±—Ä—å', 10:'–û–∫—Ç—è–±—Ä—å', 11:'–ù–æ—è–±—Ä—å', 12:'–î–µ–∫–∞–±—Ä—å'}
    for key in d:
        if mth == key:
             return d[key]
            
#–ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ —Å—Ç—Ä–æ–∫–∞–º –∏ –∑–∞–ø–∏—Å—ã–≤–∞–µ–º –≤ —Å—Ç–æ–ª–±—Ü—ã month, year 
for i, r in real_estate_data.iterrows():
    real_estate_data.loc[i, 'month'] =  month_naim(r['first_day_exposition'].month)
    real_estate_data.loc[i, 'year'] =  r['first_day_exposition'].year
    
# –ø–µ—Ä–µ–≤–æ–¥–∏–º –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–∞ year  –≤ —Ç–∏–ø int   
real_estate_data['year'] =  real_estate_data['year'].astype('int')



display(real_estate_data.head(5))
```

–î–∞, —Ö–æ—Ä–æ—à–æ, —á—Ç–æ –ø–æ–ø—Ä–æ–±–æ–≤–∞–ª–∞! –î–∞, –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ, –≤—ã–≤–æ–¥ –Ω–∞ –∞–Ω–≥–ª–∏–π—Å–∫–æ–º —è–∑—ã–∫–µ, —ç—Ç–æ –ø—Ä–æ—Å—Ç–æ –∏–Ω—Ç–µ—Ä–µ—Å–Ω—ã–µ –º–µ—Ç–æ–¥—ã, –∫–æ—Ç–æ—Ä—ã–µ –º–æ–∂–Ω–æ –∏–∑—É—á–∏—Ç—å –¥–ª—è —Å–µ–±—è –∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –≤ —Ä–∞–±–æ—Ç–µ, –µ—Å–ª–∏ –±—É–¥–µ—Ç –ø–æ—Ç—Ä–µ–±–Ω–æ—Å—Ç—å!

```python
# –°—Ç–æ–ª–±–µ—Ü "—ç—Ç–∞–∂ –∫–≤–∞—Ä—Ç–∏—Ä—ã"
# –≤–∞—Ä–∏–∞–Ω—Ç—ã ‚Äî –ø–µ—Ä–≤—ã–π, –ø–æ—Å–ª–µ–¥–Ω–∏–π, –¥—Ä—É–≥–æ–π;
def floor_naim(fl, fl_total):
    d = {1: '–ü–µ—Ä–≤—ã–π', 2:'–î—Ä—É–≥–æ–π', 3:'–ü–æ—Å–ª–µ–¥–Ω–∏–π'}
    dc = 2
    if fl == 1 and fl != fl_total:
        dc = 1
    if fl > 1 and fl != fl_total:
        dc = 2
    if fl > 1 and fl == fl_total:
        dc = 3
    return [d[dc], dc]
            
# –≤ —Å—Ç–æ–ª–±–µ—Ü floor_naim —É–∫–∞–∑—ã–≤–∞—é —Ä–µ–∑—É–ª—å—Ç–∞—Ç —É—Å–ª–æ–≤–∏—è –ø–æ –ø–æ–ª—è–º floor –∏ floors_total
for i,r in real_estate_data.iterrows():
    fl_res = floor_naim(r['floor'], r['floors_total'])
    real_estate_data.loc[i, 'floor_naim'] = fl_res[0]
    real_estate_data.loc[i, 'floor_naim_id'] = fl_res[1]

display(real_estate_data.head(5))
```

2. –ï—Å–ª–∏ —ç—Ç–∞–∂ –Ω–µ –ø–µ—Ä–≤—ã–π –∏ –Ω–µ —Ä–∞–≤–µ–Ω –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –≤—Å–µ–≥–æ —ç—Ç–∞–∂–µ–π, –∑–Ω–∞—á–∏—Ç —ç—Ç–æ —ç—Ç–∞–∂ —Å –ø–æ–º–µ—Ç–∫–æ–π "–î—Ä—É–≥–æ–π"
3. –ï—Å–ª–∏ —ç—Ç–∞–∂ –Ω–µ –ø–µ—Ä–≤—ã–π –∏ —Ä–∞–≤–µ–Ω –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –≤—Å–µ–≥–æ —ç—Ç–∞–∂–µ–π, —Ç–æ –∑–Ω–∞—á–∏—Ç —ç—Ç–æ –ø–æ—Å–ª–µ–¥–Ω–∏–π —ç—Ç–∞–∂

```python
# –†–∞—Å—á—ë—Ç —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –∂–∏–ª–æ–π –∏ –æ–±—â–µ–π –ø–ª–æ—â–∞–¥–∏
# —Å–æ–∑–¥–∞–ª–∞ –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü K, –∫–æ—Ç–æ—Ä–∞—è –∑–∞–ø–æ–ª–Ω—è–µ—Ç—Å—è –ø—É—Ç–µ–º –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –∂–∏–ª–∞—è –ø–ª–æ—â–∞–¥—å/–æ–±—â—É—é –ø–ª–æ—â–∞–¥—å
for ind, row in real_estate_data.iterrows():
    real_estate_data.loc[ind, 'K'] = row['living_area']/row['total_area']
    
# –†–∞—Å—á—ë—Ç –æ—Ç–Ω–æ—à–µ–Ω–∏–µ –ø–ª–æ—â–∞–¥–∏ –∫—É—Ö–Ω–∏ –∫ –æ–±—â–µ–π
# —Å–æ–∑–¥–∞–ª–∞ –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü K_kitchen, –∫–æ—Ç–æ—Ä–∞—è –∑–∞–ø–æ–ª–Ω—è–µ—Ç—Å—è –ø—É—Ç–µ–º –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –ø–ª–æ—â–∞–¥—å –∫—É—Ö–Ω–∏/–æ–±—â—É—é –ø–ª–æ—â–∞–¥—å
for ind, row in real_estate_data.iterrows():
    real_estate_data.loc[ind, 'K_kitchen'] = row['kitchen_area']/row['total_area']
    
display(real_estate_data.head(5))

```

—Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ –∫–æ–ª-–≤–æ –¥–Ω–µ–π –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—å—è–≤–ª–µ–Ω–∏—è —Ä–∞–≤–Ω–∞ 176 –¥–Ω–µ–π.


–ú–æ–∂–Ω–æ —Å—á–∏—Ç–∞—Ç—å, —á—Ç–æ –ø—Ä–æ–¥–∞–∂–∏ –ø—Ä–æ—à–ª–∏ –æ—á–µ–Ω—å –±—ã—Å—Ç—Ä–æ, –µ—Å–ª–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–Ω–µ–π –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä–≤–ª–µ–Ω–∏—è –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –º–µ–Ω—å—à–µ –º–µ–¥–∏–∞–Ω—ã –∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ –¥–æ–ª–≥–æ –∫–æ–≥–¥–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–Ω–µ–π –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –±–æ–ª—å—à–µ —Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–≥–æ.

```python
'''
# –°–æ–±–∏—Ä–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã–π –ø–æ—Å—á–µ—Ç –ø–æ —Ü–µ–Ω–µ 
counts = real_estate_data['last_price'].value_counts()

# –ò—Å–∫–ª—é—á–∞–µ—Ç —Å —Ç–∞–±–ª–∏—Ü—ã —Ä–µ–¥–∫–æ–≤—Å—Ç—Ä–µ—á–∞—é—â–∏–µ —Ü–µ–Ω—ã —Ç–µ, –∫–æ—Ç–æ—Ä—ã–µ –≤ —Ç–∞–±–ª–∏—Ü–µ 1 —Ä–∞–∑.
real_estate_data = real_estate_data.loc[real_estate_data['last_price'].isin(counts[counts > 1].index), :]
real_estate_data.info()
'''

#–°—Ç—Ä–æ—é –≥—Ä–∞—Ñ–∏–∫ —Ä–∞–∑–º–∞—Ö–∞  –¥–ª—è —Å—Ç–æ–ª–±—Ü–∞ ceiling_height
plt.ylim(real_estate_data['last_price'].min(), real_estate_data['last_price'].max())
last_pr = plt.boxplot(real_estate_data['last_price'])
plt.show()
# –ü–æ–ª—É—á–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –∫–≤–∞–Ω—Ç–∏–ª–∏–π Q1 –∏ Q3
Q1 = last_pr['whiskers'][0].get_ydata()[1]
Q3 = last_pr['whiskers'][1].get_ydata()[1]
print('Q1-', Q1)
print('Q3-', Q3)

# –ü–µ—Ä–µ—Å–æ–∑–¥–∞—é —Ç–∞–±–ª–∏—Ü—É —Å –Ω–æ—Ä–º–∞–ª—å–Ω—ã–º–∏ –∑–Ω–∞—á–Ω–µ–∏—è–º–∏ –≤ —Å—Ç–æ–ª–±—Ü–µ  ceiling_height –≤—Ö–æ–¥—è—â–∏–µ –≤ –æ—Ç—Ä–µ–∑–æ–∫ Q1 –∏ Q3
real_estate_data= real_estate_data.query('(last_price>=@Q1)&(last_price<=@Q3)')

real_estate_data.info()
```
